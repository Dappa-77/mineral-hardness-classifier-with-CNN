```python
import os
from icrawler.builtin import BingImageCrawler

def download_images(query, num_images):
    folder_name = query.replace(" ", "_").lower()
    
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    
    print(f"\n📁 Folder: {folder_name}")
    print(f"🔍 Downloading {num_images} images for '{query}'...")
    
    crawler = BingImageCrawler(storage={'root_dir': folder_name})
    
    crawler.crawl(
        keyword=query,
        max_num=num_images
    )
    
    print(f"✅ Finished: {query}\n")


if __name__ == "__main__":
    # List of Mohs minerals (optimized search queries)
    minerals = [
        "talc mineral specimen",
        "gypsum mineral specimen",
        "calcite mineral specimen",
        "fluorite crystal mineral",
        "apatite mineral rock",
        "orthoclase feldspar mineral",
        "quartz crystal mineral",
        "topaz crystal stone",
        "corundum mineral ruby sapphire",
        "diamond rough crystal"
    ]
    
    num_images = int(input("Enter number of images per mineral: "))
    
    for mineral in minerals:
        download_images(mineral, num_images)
    
    print("\n🎉 All minerals downloaded successfully!")
```

```python
import os
from PIL import Image
import imagehash
from tqdm import tqdm

# SETTINGS
IMAGE_SIZE = (224, 224)  # For ResNet
HASH_SIZE = 8            # Controls duplicate detection sensitivity


def clean_folder(folder_path):
    print(f"\n🧹 Cleaning folder: {folder_path}")
    
    hashes = {}
    removed = 0
    processed = 0

    for filename in tqdm(os.listdir(folder_path)):
        file_path = os.path.join(folder_path, filename)
        
        try:
            with Image.open(file_path) as img:
                # Convert to RGB (important)
                img = img.convert("RGB")
                
                # Resize image
                img = img.resize(IMAGE_SIZE)
                
                # Generate hash
                img_hash = imagehash.average_hash(img, HASH_SIZE)
                
                # Check duplicates
                if img_hash in hashes:
                    os.remove(file_path)
                    removed += 1
                else:
                    hashes[img_hash] = file_path
                    # Save resized image (overwrite)
                    img.save(file_path, "JPEG", quality=90)
                    processed += 1

        except Exception as e:
            # Remove corrupted images
            os.remove(file_path)
            removed += 1

    print(f"✅ Done: {processed} images kept, {removed} removed")


def clean_dataset(base_path):
    print("\n🚀 Starting dataset cleaning...\n")
    
    for folder in os.listdir(base_path):
        folder_path = os.path.join(base_path, folder)
        
        if os.path.isdir(folder_path):
            clean_folder(folder_path)
    
    print("\n🎉 Dataset cleaned successfully!")


if __name__ == "__main__":
    base_path = input("Enter dataset folder path: ")
    clean_dataset(base_path)
```

```python
import os
import shutil
import random

def split_dataset(base_path, output_path, train_ratio=0.7, val_ratio=0.2, test_ratio=0.1):
    print("\n🚀 Splitting dataset...\n")

    classes = [d for d in os.listdir(base_path) if os.path.isdir(os.path.join(base_path, d))]

    for cls in classes:
        class_path = os.path.join(base_path, cls)
        images = os.listdir(class_path)

        random.shuffle(images)

        total = len(images)
        train_end = int(train_ratio * total)
        val_end = train_end + int(val_ratio * total)

        train_imgs = images[:train_end]
        val_imgs = images[train_end:val_end]
        test_imgs = images[val_end:]

        for split, split_imgs in zip(['train', 'val', 'test'], [train_imgs, val_imgs, test_imgs]):
            split_dir = os.path.join(output_path, split, cls)
            os.makedirs(split_dir, exist_ok=True)

            for img in split_imgs:
                src = os.path.join(class_path, img)
                dst = os.path.join(split_dir, img)

                try:
                    shutil.copy2(src, dst)
                except Exception:
                    continue

        print(f"✅ {cls}: {total} images split")

    print("\n🎉 Dataset split complete!")


if __name__ == "__main__":
    base_path = input("Enter original dataset path: ").strip().strip('"')
    output_path = input("Enter output folder path: ").strip().strip('"')

    split_dataset(base_path, output_path)
```
